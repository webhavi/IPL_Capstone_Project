```python
HomeTeam = df['Home Team'].value_counts()
sns.barplot(y=HomeTeam.index , x=HomeTeam.values)
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("count")
plt.title("Home Team Matches")
plt.show()
```
